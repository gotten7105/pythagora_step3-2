guests Tableにデータ追加

```python
from sqlalchemy import create_engine, text
from datetime import datetime

now = datetime.now()

# データベースエンジンの作成
engine = create_engine('sqlite:///Cheers_DB.db')

# データベースに接続
with engine.connect() as connection:
    # トランザクションの開始
    trans = connection.begin()

    try:
        # データの挿入
        connection.execute(text("""
            INSERT INTO guests (
                guest_id,
                guest_name,
                guest_class,
                user_id,
                guest_category1,
                guest_category2,
                guest_category3,
                guest_memo,
                last_update
            ) VALUES (
                :guest_id,
                :guest_name,
                :guest_class,
                :user_id,
                :guest_category1,
                :guest_category2,
                :guest_category3,
                :guest_memo,
                :last_update
            )"""),
            [{"guest_id": 1,
              "guest_name": "濱田隼斗",
              "guest_class": "CEO",
              "user_id": 3,
              "guest_category1": "鮨",
              "guest_category2": "焼肉",
              "guest_category3": " ",
              "guest_memo": "飲み会大好きです！",
              "last_update": now
             },
             {"guest_id": 2,
              "guest_name": "鍵冨康太郎",
              "guest_class": "CFO",
              "user_id": 3,
              "guest_category1": " ",
              "guest_category2": " ",
              "guest_category3": " ",
              "guest_memo": "ローマが好きなので、きっとイタリアン好きだと思う",
              "last_update": now
             }]
        )

        # トランザクションのコミット
        trans.commit()
        print("データの挿入に成功しました。")
    except Exception as e:
        # 何か問題が発生した場合、トランザクションをロールバック
        trans.rollback()
        print(f"エラーが発生しました: {e}")
```
