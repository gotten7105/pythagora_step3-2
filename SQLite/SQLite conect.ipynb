```python
import json
from sqlalchemy import create_engine, text

# データベースエンジンの作成
engine = create_engine('sqlite:///Cheers_DB.db')   #///以降に各自Cheers_DB.dbの置いた場所を設定下さい！
connection = engine.connect()

# SQLクエリの実行
result = connection.execute(text("SELECT * FROM users"))

# 結果を辞書のリストとして取得
rows = [{column: value for column, value in zip(result.keys(), row)} for row in result]

# JSON形式に変換（非ASCII文字をエスケープしない）
json_result = json.dumps(rows, ensure_ascii=False, indent=4)

# JSON形式の結果を出力
print(json_result)

# コネクションのクローズ
connection.close()



```

Tableの上書き

```python
from sqlalchemy import create_engine, text
from datetime import datetime

# データベースエンジンの作成
engine = create_engine('sqlite:///Cheers_DB.db')

# データベースに接続
with engine.connect() as connection:
    # トランザクションの開始
    trans = connection.begin()

    try:
        # 特定のユーザーのデータを更新
        connection.execute(text("""
            UPDATE participant_information
            SET event_comment = :event_comment
            WHERE user_id = :user_id
        """), {
            "event_comment": "剛さん幹事ありがとう！迷惑おかけしました",
            "user_id": 4  # 更新したいユーザーのID
        })

        # トランザクションのコミット
        trans.commit()
        print("データの更新に成功しました。")
    except Exception as e:
        # エラーが発生した場合、トランザクションをロールバック
        trans.rollback()
        print(f"更新に失敗しました: {e}")
```

```python
from sqlalchemy import create_engine, text
from datetime import datetime

now = datetime.now()

# データベースエンジンの作成
engine = create_engine('sqlite:///Cheers_DB.db')

# データベースに接続
with engine.connect() as connection:
    # トランザクションの開始
    trans = connection.begin()

    try:
        # DataFrame df の各行に対してデータの挿入
        for index, row in df.iterrows():
            connection.execute(text("""
                INSERT INTO restaurants (
                    restaurant_id,
                    restaurant_name,
                    genre,
                    average_charge,
                    restaurant_url,
                    restaurant_image,
                    restaurant_public_evaluation,
                    restaurant_address,
                    regular_holiday,
                    open_time,
                    last_update
                ) VALUES (
                    :restaurant_id,
                    :restaurant_name,
                    :genre,
                    :average_charge,
                    :restaurant_url,
                    :restaurant_image,
                    :restaurant_public_evaluation,
                    :restaurant_address,
                    :regular_holiday,
                    :open_time,
                    :last_update
                )"""),
                {
                    "restaurant_id": row['restaurant_id'],
                    "restaurant_name": row['restaurant_name'],
                    "genre": row['genre'],
                    "average_charge": row['average_charge'],
                    "restaurant_url": row['restaurant_url'],
                    "restaurant_image": row['restaurant_image'],
                    "restaurant_public_evaluation": row['restaurant_public_evaluation'],
                    "restaurant_address": row['restaurant_address'],
                    "regular_holiday": row['regular_holiday'],
                    "open_time": row['open_time'],
                    "last_update": now
                }
            )

        # トランザクションのコミット
        trans.commit()
        print("データの挿入に成功しました。")
    except Exception as e:
        # 何か問題が発生した場合、トランザクションをロールバック
        trans.rollback()
        print(f"エラーが発生しました: {e}")
```
